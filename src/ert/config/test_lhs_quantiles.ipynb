```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

num_realizations = 1000
seed = 123
rng = np.random.default_rng(seed)
lhs_values = []
normal_values = []

for realization in range(num_realizations):
    # Latin hypercube sampling
    sampler = sp.stats.qmc.LatinHypercube(1, rng=rng)
    quantile = sampler.random(num_realizations)
    lhs_value = sp.stats.norm(0, 1).ppf(quantile[realization][0])
    normal_value = rng.standard_normal(1)
    lhs_values.append(lhs_value)
    normal_values.append(normal_value)

plt.hist(np.array(lhs_values), bins=30, density=True)
plt.hist(np.array(normal_values), bins=30, color="g", alpha=0.5, density=True)

# Test which converges faster to normal dist??

x = np.linspace(-4, 4, 100)
plt.plot(x, sp.stats.norm.pdf(x))
```
